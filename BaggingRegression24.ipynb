```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv("/content/BostonHousing.csv")
data.head()
```

```python
x = data.drop("medv",axis=1)
y = data["medv"]
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size =0.2,random_state=42)
```

```python
from sklearn.ensemble import BaggingRegressor
br = BaggingRegressor()
br.fit(x_train,y_train)
```

```python
y_pred_train = br.predict(x_train)
y_pred_test = br.predict(x_test)
```

```python
from sklearn.metrics import mean_absolute_error
print("Train")
print("mean_absolute_error is :",mean_absolute_error(y_train,y_pred_train))
print("Test")
print("mean_absolute_error is :",mean_absolute_error(y_test,y_pred_test))
```

```python
from sklearn.metrics import mean_absolute_error,mean_absolute_percentage_error,mean_squared_error,r2_score
print("Train")
print("The mean_absolute_percentage_error of model is : ",mean_absolute_percentage_error(y_train,y_pred_train))
print("The mean_squared_error of model is : ",mean_squared_error(y_train,y_pred_train))
print("The r2_score of model is : ",r2_score(y_train,y_pred_train))
print("The root_mean_square_error of model is :\n ",np.sqrt(y_train,y_pred_train))
     
```

```python
from sklearn.metrics import mean_absolute_error,mean_absolute_percentage_error,mean_squared_error,r2_score
print("Test")
print("The mean_absolute_percentage_error of model is : ",mean_absolute_percentage_error(y_test,y_pred_test))
print("The mean_squared_error of model is : ",mean_squared_error(y_test,y_pred_test))
print("The r2_score of model is : ",r2_score(y_test,y_pred_test))
print("The root_mean_square_error of model is :\n ",np.sqrt(y_test,y_pred_test))
```
