POLYNOMIAL REGRESSION 


```python
# Importing required libraries
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
x1 = np.random.rand(300)+4*(np.random.rand(300))+2
x2 = 5*(np.random.rand(300))+ 4
# y = 2*x1sq + 3*x2sq +12
y = 2*x1**2 + 3*x2**2 +12
print(y.shape)
```

```python
plt.scatter(x2,y)
```

```python
# Creating data frame
data = pd.DataFrame({'x1':x1,'x2':x2,'y_actual':y})
data.head()
```

```python
# Inserting x1 square , x2 square , x1*x2
data['x1_sq'] = data['x1']*data['x1']
data['x2_sq'] = data['x2']*data['x2']
data['x1_x2'] = data['x1']*data['x2']
data.head()
```

```python
input = data.drop("y_actual", axis = 1)
output = data['y_actual']
```

```python
# Polynomial Regression Model 
from sklearn.linear_model import LinearRegression 
poly_reg = LinearRegression()
poly_reg.fit(input,output)
```

```python
# Prediction using Model 
y_pred = poly_reg.predict(input)
print("The intercept is: ",poly_reg.intercept_)
print("The coefficient is: ",poly_reg.coef_)
```

```python
plt.scatter(y_pred,output)
```

```python
# Calculation of how the model performed using Metrics 
from sklearn.metrics import mean_absolute_error,mean_absolute_percentage_error,mean_squared_error,r2_score
print("The mean_absolute_error of the model is: ",mean_absolute_error(output,y_pred))
print("The mean_absolute_percentage_error of the model is: ",mean_absolute_percentage_error(output,y_pred))
print("The mean_squared_error of the model is: ",mean_squared_error(output,y_pred))
print("The r2_score of the model is: ",r2_score(output,y_pred))
```
