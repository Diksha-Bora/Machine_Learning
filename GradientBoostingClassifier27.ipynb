```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
```

```python
from sklearn.datasets import load_iris
iris = load_iris()
data = pd.DataFrame(iris.data, columns = iris.feature_names)
data["target"] = iris.target
data.columns = data.columns.str.capitalize()
data.head()
```

```python
x = data.drop("Target",axis=1)
y = data["Target"]
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=42,stratify=y)
```

```python
from sklearn.tree import DecisionTreeClassifier 
dtc = DecisionTreeClassifier()
dtc.fit(x_train,y_train)
```

```python
y_pred_train = dtc.predict(x_train)
y_pred_test = dtc.predict(x_test)
```

```python
from sklearn.metrics import classification_report 
print("TRAIN")
print("The classification_report is : \n",classification_report(y_train,y_pred_train))
print("TEST")
print("The classification_report is : \n",classification_report(y_test,y_pred_test))
```

GRADIENT BOOSTING CLASSIFIER WITH NO HYPERPARAMETER 

```python
from sklearn.ensemble import GradientBoostingClassifier 
gbc = GradientBoostingClassifier()
gbc.fit(x_train,y_train)
```

```python
y_pred_train1 = gbc.predict(x_train)
y_pred_test1 = gbc.predict(x_test)
```

```python
from sklearn.metrics import classification_report 
print("TRAIN")
print("The classification_report is : \n",classification_report(y_train,y_pred_train1))
print("TEST")
print("The classification_report is : \n",classification_report(y_test,y_pred_test1))
```

GRADIENT BOOSTING CLASSIFIER WITH HYPERPARAMETERS 

```python
from sklearn.ensemble import GradientBoostingClassifier 
gbch = GradientBoostingClassifier(learning_rate=0.2,n_estimators=150,warm_start=True,verbose=True,criterion="friedman_mse")
gbch.fit(x_train,y_train)
```

```python
y_pred_train2 = gbch.predict(x_train)
y_pred_test2 = gbch.predict(x_test)
```

```python
from sklearn.metrics import classification_report 
print("TRAIN")
print("The classification_report is : \n",classification_report(y_train,y_pred_train2))
print("TEST")
print("The classification_report is : \n",classification_report(y_test,y_pred_test2))
```
